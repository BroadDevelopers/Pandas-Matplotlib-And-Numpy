```python
'''
we want to replace all the -99999 with na becausee they represent
error/nan values
'''
'''
Instead of having only -99999 we have -88888
we replace the -88888,-99999,0 and by putting them into a list
but weather accepts zero-0 as a value
so we want to replace values based on a specific value
'''
#new_df = df.replace([-99999,-88888], np.NaN)
'''
we want to replace values based on a specefic column using
dictionary
replace takes in two parameteres replace(value we want to replace, with np.NaN)
'''
new_df = df.replace({
    'temperature': -99999,
    'windspeed':-99999,
    'event': '0'
}, np.NaN)
```

```python
new_df
```

```python
'''
lets say now where we had no event we want to replace with sunny
and -99999 with np.NaN
we can do this using a fictionary for mapping
'''
new_df = df.replace({
    -99999:np.NaN,
    'No Event': 'Sunny'
})
new_df
```

```python
'''
we want to remove the measures(mph, C, F) only from our values
the best way to do this is to use a regx-regular expression
regx are used to detect patterns
'''
#new_df = df.replace('[A-Za-z]','',regex=True) this will delete the event values
#so we want to erase based on column using a dictionary
new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```
