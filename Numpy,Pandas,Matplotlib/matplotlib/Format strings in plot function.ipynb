```python
"""
Format string are used to control visual aspect of your 2D plot
such as color,marker style,linestyle....etc
"""
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
"""
we can control visual properties of this chart by specifying format strings
as 3rd arguement
g+=g means green, plus means marker
"""
plt.plot(x,y,'rD')
```

```python
"""
if you dont like format strings you can provide individual arguement
rgb colors can also be used
marker size can be controlled using markersize
"""
plt.plot(x,y,color='green',marker='D',linestyle='',markersize=12)
```

```python
"""
one more property that plot has is alpha
alpha controls transparency
alpha is between 0-1
1-
"""
plt.plot(x,y,color='green',alpha=0.2)
```
