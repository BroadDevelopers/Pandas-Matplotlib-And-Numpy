```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vls=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill','Car','Other Utilities']
```

```python
plt.axis("equal")#makes pie chart look like a perfect cirlce
plt.pie(exp_vls,labels=exp_labels,radius=1.5,autopct='%0.3f%%',shadow=True,explode=[0,0.1,0,0,0])
plt.show()#gets rid of the text
#radius increases chart size
#autopct shows percentage of chart-0.2 shows number of decimal point
#shadow
#explode takes out the pies
#explode takes in array of number of values=number of pies

#saaving matplotlib chart to a file
plt.savefig("piechart.png")
```
